Run Simotaneous Pose and Geometry Estimation

```python
from helper import pose_to_transform
from pose_config import test_poses
import torch
from simo_est import read_preprocessed, stage_1_opt, stage_2_opt
from simo_est_pred import pred_on_test_init, pred_on_test

test_poses = pose_to_transform(torch.tensor(test_poses).clone()).float()
train_dir = f"{data_dir}/{tool}/train"

poses, pts_tor, rgb_tor, masks = read_preprocessed(data_dir, tool, num_images)
"""
Notice pred function when plot img/projection on matplotlib plot from the upper origin, so the
prediction looks upside-down
"""
print("done pred train")
while (True):
    try:
        stage_1_res = stage_1_opt(poses, pts_tor, masks, tool, num_images)
        T, T_x_s_aug, scale_t, trans, poses_t, eef_poses, scale_t, gt_inview, gt_bool = stage_1_res
        # pred_on_train(poses_t, eef_poses, pts_tor, rgb_tor, num_used, T, scale_t, train_dir)
        print("Visualizing test set result from stage 1 optimization result")
        pred_on_test_init(T, scale_t, test_poses, poses_t, eef_poses, pts_tor, rgb_tor, num_images, 
                    data_dir, tool, save_pred_imgs=True)
        
        R_c, trans_c, scale_t_c = stage_2_opt(T_x_s_aug, scale_t, trans, poses_t, eef_poses, 
                                            pts_tor, gt_inview, gt_bool, num_images)
        break
    except:
        print("")
        user_input = input("Gradient exceeding, retrying? (0 for quit, 1 for retrying)")
        if user_input == '0':
            print("Exiting")
            break
        else:
            print("Retrying...")
print("Visualizing test set result from stage 2 optimization result")
pred_on_test(R_c, trans_c, scale_t_c, 
                test_poses, poses_t, eef_poses, pts_tor, rgb_tor, num_images, 
                data_dir, tool, save_pred_imgs=True)
```
